**Problem Introduction**
1. Relationship between this experiment and the original with Fisher and Bristol <br>

Both experiments are similar in the sense that they want to test if the person can actually distinguish the taste of tea if the milk is poured in first or after, or they only guess it correctly by chance <br>

2. Statements of the Null Hypothesis and Alternative hypothesis
Null hypothesis: There is no difference in taste when the milk is poured into the cup before or after the tea. Alternative hypothesis: There is a difference in taste when the milk is poured into the cup before or after the tea.




**Quantitative Analysis** <br>
Methodology Code and Explanations <br>
(if needed/optional) Supporting Visualizations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Set parameters
n = 80  # total number of students
k = 49  # number of students who guessed correctly
observed_proportion = k / n
alpha = 0.05  # significance level

# Bootstrapping
number_of_simulations = 10000
bootstrapped_proportions = np.zeros(number_of_simulations)

# Create an array with the outcomes: 1 for correct guesses, 0 for incorrect guesses
data = np.array([1] * k + [0] * (n - k))

# Perform bootstrapping
for i in range(number_of_simulations):
    # Sample with replacement from the original data
    sample = np.random.choice(data, size=n, replace=True)
    # Calculate the proportion of correct guesses in the bootstrap sample
    bootstrapped_proportions[i] = np.mean(sample)

# Calculate the p-value
# Count how many bootstrap proportions are greater than or equal to the observed proportion
p_value = np.mean(bootstrapped_proportions >= observed_proportion)

# Output results
print(f"Observed Proportion: {observed_proportion}")
print(f"P-value: {p_value}")

# Visualization of the bootstrap distribution
plt.hist(bootstrapped_proportions, bins=30, color='lightblue', edgecolor='black')
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f'Observed Proportion: {observed_proportion}')
plt.title('Bootstrapped Proportions Distribution')
plt.xlabel('Proportion')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

**Findings and Discussion** <br>
Conclusion regarding the Null Hypothesis

The p-value is 0.5568, which is a high value, meaning that 55.68% of the time, the proportion of the students guessing the difference in the tea correctly is by chance, not because they can tell the difference. Since this p-value is greater than the observed statistic (0.6125), and it is much larger than the typical significance level (Œ± = 0.05) we cannot reject the null hypothesis that there is no difference in the tea pouring order, as the evidence is not strong or extreme enough to suggest otherwise.
